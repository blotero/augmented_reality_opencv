## Apply generated markers

```python
input_path = "resources/input_1.jpeg"
source_path = "resources/source_1.jpeg"
image = cv2.imread(input_path)
image = imutils.resize(image, width=600)
source = cv2.imread(source_path)
plt.figure(figsize=(8,5))
plt.subplot(121)
plt.imshow(image)
plt.title("Input image")
plt.subplot(122)
plt.imshow(source)
plt.title("Source image")
```

```python
#Import transformer from internal source utils
from src.lib.core.augtransformer import AugTransformer

autrf = AugTransformer(source)
output = autrf.transform(image)
```

```python
plt.figure(figsize=(15,5))
plt.subplot(131)
plt.imshow(image)
plt.subplot(132)
plt.imshow(source)
plt.subplot(133)
plt.imshow(output)
plt.imsave("resources/output.png",output)
```
